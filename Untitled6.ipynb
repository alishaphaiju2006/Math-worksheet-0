```python
#Problem - 4: Matrix Operations:

import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1
A_inv = np.linalg.inv(A)
print("A * A^-1:\n", A @ A_inv)

# 2
print("AB:\n", A @ B)
print("BA:\n", B @ A)

# 3
print("(AB)^T:\n", (A @ B).T)
print("B^T A^T:\n", B.T @ A.T)

# Solve linear system AX = B

A2 = np.array([[2, -3, 1],
               [1, -1, 2],
               [3, 1, -1]])

B2 = np.array([-1, -3, 9])

# Solve using inverse
A2_inv = np.linalg.inv(A2)
X = A2_inv @ B2
print("Solution using inverse:\n", X)

# Solve using np.linalg.solve
X2 = np.linalg.solve(A2, B2)
print("Solution using np.linalg.solve:\n", X2)
```
